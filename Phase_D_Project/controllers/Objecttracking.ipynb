```python
import numpy as np
import cv2 as cv
import argparse
import matplotlib.pyplot as plt # Matplotlib library for plotting

cap = cv.VideoCapture('../RobotTrackingFeatures.mp4')
# take first frame of the video
ret,frame = cap.read()
plt.figure(figsize = (18, 10))
plt.imshow(frame)
plt.show()

# setup initial location of window
x, y, w, h = 125, 90, 100, 100 # simply hardcoded the values of object in first frame
track_window = (x, y, w, h)
frame = cv.cvtColor(frame, cv.COLOR_BGR2RGB)

# set up the ROI for tracking object
roi = frame[y:y+h, x:x+w]
hsv_roi =  cv.cvtColor(roi, cv.COLOR_BGR2HSV)
plt.imshow(roi)
plt.show()
mask = cv.inRange(hsv_roi, (15,0,89), (140,186,195)) #hsv values of object
roi_hist = cv.calcHist([hsv_roi],[0],mask,[180],[0,180])
cv.normalize(roi_hist,roi_hist,0,255,cv.NORM_MINMAX)
# Setup the termination criteria, either 10 iteration or move by atleast 1 pt
term_crit = ( cv.TERM_CRITERIA_EPS | cv.TERM_CRITERIA_COUNT, 60, 1 )
result = cv.VideoWriter('../MotionTrackingResult.avi',cv.VideoWriter_fourcc(*'MJPG'), 40, (1920,1080))

while(1):
    ret, frame = cap.read()
    if ret == True:
        hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
        dst = cv.calcBackProject([hsv],[0],roi_hist,[0,180],1)
        
        # apply camshift to get the new location
        ret, track_window = cv.CamShift(dst, track_window, term_crit)
        
        X = track_window[0]
        Y = track_window[1]
        text = str([X,Y])
        cv.putText(frame, text,(track_window[0]-20, track_window[1]-20),cv.FONT_HERSHEY_SIMPLEX,0.8,(0, 255, 0),2,cv.LINE_4)
#         cv.putText(frame, '(x,y)',(track_window[0], track_window[1]-60),cv.FONT_HERSHEY_SIMPLEX,0.8,(0, 255, 0),2,cv.LINE_4)
        # Draw it on image
        pts = cv.boxPoints(ret)
        pts = np.int0(pts)
        img2 = cv.polylines(frame,[pts],True, (0,255,0),2)
#         height = int(img2.shape[0]/2)
#         width = int(img2.shape[1]/2)
#          print(height)
#         imgToShow = cv.resize(img2, (1920,108))
#         cv.imshow('hi',imgToShow)
        result.write(img2)
        cv.imshow('img2',img2)
        if cv.waitKey(1) == ord('q'):
            break
    else:
        break
        
cv.destroyAllWindows()
```
